Q1. Apply regular expression for form validation. Create your domain-form using Tkinter
Module.
➔ Form should contain Text box [For Name, Email Id, Phone number], Dropdown [for
Gender], Spinbox [for Year/DoB] and other necessary widgets required for your
domain.
➔ Validate Your Name, Email Id, Phone number in the form.

```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

# Function to validate the form fields
def validate_form():
    name = name_entry.get()
    email = email_entry.get()
    age = age_entry.get()
    phone = phone_entry.get()  # Get the phone number
    place = place_combobox.get()
    destination = destination_combobox.get()
    gender = gender_combobox.get()  # Get the selected gender
    dob = dob_spinbox.get()  # Get the date of birth

    # Check if any field is empty
    if not name or not email or not age or not phone or not place or not destination or not gender or not dob:
        messagebox.showerror("Error", "All fields must be filled.")
        return

    # Check if the name contains numbers
    if any(char.isdigit() for char in name):
        messagebox.showerror("Error", "Name should not contain numbers.")
        return

    # Check if the email follows the format
    if not "@" in email or not "." in email:
        messagebox.showerror("Error", "Invalid email format.")
        return

    # Check if age is above 18
    try:
        age = int(age)
        if age <= 18:
            messagebox.showerror("Error", "Age must be above 18.")
            return
    except ValueError:
        messagebox.showerror("Error", "Age must be a valid number.")
        return

    # Check if phone number is numeric and has 10 digits
    if not phone.isdigit() or len(phone) != 10:
        messagebox.showerror("Error", "Invalid phone number format. It should be a 10-digit number.")
        return

    # If all validations pass, display success message
    messagebox.showinfo("Success", "Form submitted successfully!")

# Create the main window
root = tk.Tk()
root.title("Tourism Form")

# Create a frame for the form
frame = tk.Frame(root)
frame.pack(padx=20, pady=20)

# Create form elements with spaces
tk.Label(frame, text="Tourist Details", font=("Arial", 16)).grid(row=0, columnspan=2, pady=10)

tk.Label(frame, text="Name:").grid(row=1, column=0, pady=5)
name_entry = tk.Entry(frame)
name_entry.grid(row=1, column=1, pady=5)

tk.Label(frame, text="Email:").grid(row=2, column=0, pady=5)
email_entry = tk.Entry(frame)
email_entry.grid(row=2, column=1, pady=5)

tk.Label(frame, text="Gender:").grid(row=3, column=0, pady=5)
genders = ["Male", "Female", "Other"]
gender_combobox = ttk.Combobox(frame, values=genders)
gender_combobox.grid(row=3, column=1, pady=5)

tk.Label(frame, text="Age:").grid(row=4, column=0, pady=5)
age_entry = tk.Entry(frame)
age_entry.grid(row=4, column=1, pady=5)

tk.Label(frame, text="Phone Number:").grid(row=5, column=0, pady=5)
phone_entry = tk.Entry(frame)
phone_entry.grid(row=5, column=1, pady=5)

tk.Label(frame, text="Place:").grid(row=6, column=0, pady=5)
places = ["Select", "Beach", "Mountain", "City", "Countryside"]
place_combobox = ttk.Combobox(frame, values=places)
place_combobox.grid(row=6, column=1, pady=5)
place_combobox.set(places[0])

tk.Label(frame, text="Destination:").grid(row=7, column=0, pady=5)
destinations = ["Select", "Paris", "Maldives", "Thailand", "Bali"]
destination_combobox = ttk.Combobox(frame, values=destinations)
destination_combobox.grid(row=7, column=1, pady=5)
destination_combobox.set(destinations[0])

tk.Label(frame, text="Date of Birth:").grid(row=8, column=0, pady=5)
dob_spinbox = ttk.Spinbox(frame, from_=1900, to=2023)
dob_spinbox.grid(row=8, column=1, pady=5)

# Create a Submit button
submit_button = tk.Button(frame, text="Submit", command=validate_form)
submit_button.grid(row=9, columnspan=2, pady=10)

# Start the Tkinter main loop
root.mainloop()
```
