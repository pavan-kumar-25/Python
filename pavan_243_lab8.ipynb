```python
import matplotlib.pyplot as plt
import seaborn as sns
gender_count = df['Sex'].value_counts()
plt.bar(gender_count.index, gender_count.values)
plt.title('gender of passenger')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sur_count = df['Survived'].value_counts()
plt.bar(sur_count.index, sur_count.values)
plt.title('survival count')
plt.xlabel('Surival')
plt.ylabel('Count')
plt.show()
```

```python
age_count = df['Age'].value_counts()
plt.bar(age_count.index, age_count.values)
plt.title('age of passenger')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(df['Age'], df['Fare'])
plt.title('Age and fare comapision')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
